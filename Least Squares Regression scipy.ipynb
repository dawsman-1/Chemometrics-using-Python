```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
from sklearn.metrics import r2_score
```

## Least Squares Regression using SciPy

Linear regression analyzes the relationship between two variables, x and y. In general, the goal of linear regression is to find the line that best predicts y from x. This is achieved by finding the line that minimizes the sum of the squares of the vertical distances of the points from the line – the method is commonly called ‘least squares’. The Least squares method of line fitting assumes (i) that there is a linear relationship between y and x, defined by an equation y = a + bx (with a data set of n points); (ii) all errors are in the y direction; and (iii) errors are constant (absolute).

It is important to notice that the line of regression of x on y is not the same line as y on x, except in the highly unlikely case that the correlation coefficient is 1 exactly (Miller and Miller, 1993). Therefore depending on the type of data it may or may not be essential to consider the x on y fit, and errors that occur in the x-direction.

In this case, a table of experimental data (Table 1) has been given: concentration (x) and absorbance (y):

```python
# Read data

conc_abs = pd.read_csv("Data/least_squares_data.csv")
print(conc_abs)
```

### Least squares fit assuming only y has errors

```python
# Define linear function

def func(x, a, b):
    y = a*x + b
    return y

# Fit line and print slope and intercept

alpha = optimize.curve_fit(func, 
                           xdata = conc_abs["Conc (x)"], 
                           ydata = conc_abs["Abs (y)"])[0]
print(alpha)
```

```python
# Calculate coefficient of determination (R2)

R2 = r2_score(conc_abs["Abs (y)"],
              alpha[0]*conc_abs["Conc (x)"] + alpha[1])
print(R2)
```

```python
# Plot data

plt.style.use('seaborn-poster')
plt.figure(figsize = (10,8))
plt.plot(conc_abs["Conc (x)"], 
         conc_abs["Abs (y)"], 
         'b.')

# Plot linear regression function

plt.plot(conc_abs["Conc (x)"], 
         alpha[0]*conc_abs["Conc (x)"] + alpha[1], 
         'r', 
         label='y={:.2f}x+{:.2f}; $R^2$={:.2f}'.format(alpha[0], 
                                                       alpha[1], 
                                                       R2))
plt.xlabel('Conc')
plt.ylabel('Abs')
plt.legend(fontsize=15)
plt.show()
```

### Linear regression on non-linear data

```python
# Read data

non_linear_conc_abs = pd.read_csv("Data/non_linear_data.csv")
print(non_linear_conc_abs)
```

```python
# Fit line and print slope and intercept

alpha = optimize.curve_fit(func,
                           xdata = non_linear_conc_abs["Conc (x)"],
                           ydata = non_linear_conc_abs["Abs (y)"])[0]
print(alpha)
```

```python
# Calculate coefficient of determination (R2)

R2 = r2_score(non_linear_conc_abs["Abs (y)"],
              alpha[0]*non_linear_conc_abs["Conc (x)"] + alpha[1])
print(R2)
```

```python
# Plot data

plt.figure(figsize = (10,8))
plt.plot(non_linear_conc_abs["Conc (x)"], 
         non_linear_conc_abs["Abs (y)"], 'b.')

# Plot linear regression function

plt.plot(non_linear_conc_abs["Conc (x)"], 
         alpha[0]*non_linear_conc_abs["Conc (x)"] + alpha[1], 
         'r', 
         label='y={:.3f}x+{:.3f}; $R^2$={:.3f}'.format(alpha[0], 
                                                       alpha[1], 
                                                       R2))
plt.xlabel('Conc')
plt.ylabel('Abs')
plt.legend(fontsize=15)
plt.show()
```

### Non-linear regression

#### 2nd order polynomial

```python
# Define 2nd order polynomial function

def func(x, a, b, c):
    y = a*(x**2) + b*x + c
    return y

# Fit line and print coefficients

alpha = optimize.curve_fit(func, xdata = non_linear_conc_abs["Conc (x)"],
                           ydata = non_linear_conc_abs["Abs (y)"])[0]
print(alpha)
```

```python
# Calculate coefficient of determination (R2)

R2 = r2_score(non_linear_conc_abs["Abs (y)"],
              alpha[0]*(non_linear_conc_abs["Conc (x)"])**2 \
              + (alpha[1]*non_linear_conc_abs["Conc (x)"]) + alpha[2])
print(R2)
```

```python
# Plot data

plt.figure(figsize = (10,8))
plt.plot(non_linear_conc_abs["Conc (x)"], 
         non_linear_conc_abs["Abs (y)"], 
         'b.')

# Plot 2nd order polynomial function

x = np.linspace(0, 100, 50)

plt.plot(x, 
         alpha[0]*(x**2) \
         + (alpha[1]*x) + alpha[2], 
         'r', 
         label='y={:.3e}$x^2$+{:.3e}x+{:.4f}; $R^2$={:.4f}'.format(alpha[0], 
                                                                   alpha[1], 
                                                                   alpha[2], 
                                                                   R2))
plt.xlabel('Conc')
plt.ylabel('Abs')
plt.legend(fontsize=15)
plt.show()
```

#### 3rd order polynomial

```python
# Define 3rd order polynomial function

def func(x, a, b, c, d):
    y = a*(x**3) + b*(x**2) + c*x + d
    return y

# Fit line and print coefficients

alpha = optimize.curve_fit(func,
                           xdata = non_linear_conc_abs["Conc (x)"],
                           ydata = non_linear_conc_abs["Abs (y)"])[0]
print(alpha)
```

```python
# Calculate coefficient of determination (R2)

R2 = r2_score(non_linear_conc_abs["Abs (y)"],
              (alpha[0]*(non_linear_conc_abs["Conc (x)"])**3) \
              + (alpha[1]*(non_linear_conc_abs["Conc (x)"])**2) \
              + (alpha[2]*non_linear_conc_abs["Conc (x)"]) + alpha[3])
print(R2)
```

```python
# Plot data

plt.figure(figsize = (10,8))
plt.plot(non_linear_conc_abs["Conc (x)"], 
         non_linear_conc_abs["Abs (y)"], 
         'b.')

# Plot 3rd order polynomial function

x = np.linspace(0, 100, 50)

plt.plot(x,
         alpha[0]*(x**3) + \
         alpha[1]*(x**2) + \
         (alpha[2]*x) + alpha[3], 
         'r', 
         label='y={:.3e}$x^3$+{:.3e}$x^2$+{:.3f}x+{:.3f}; $R^2$={:.3f}'.format(alpha[0],
                                                                               alpha[1],
                                                                               alpha[2],
                                                                               alpha[3],
                                                                               R2))
plt.xlabel('Conc')
plt.ylabel('Abs')
plt.legend(fontsize=15)
plt.show()
```
